## 57F07-Gal4

```python
#main_df

path = 'C:/Users/Acer/Dropbox/WaveAnalysis/57F07-Gal4/'
#path = 'C:/Users/varshinig/Dropbox/WaveAnalysis/57F07-Gal4/'

firing_data = []
segmentLoc_data = []
waveDuration = []
segmentsCovered = []
waveDirection = []
sorted_firing_data = []
sorted_segmentsCovered = []
Amplitude = []
for file in os.listdir(path):
    f = pd.read_excel(path+file)
    
    #print(file)
    #print(f.any)
    id=0
    # Firing times and segment locations of all animals
    for index, row in f.iterrows():
        if index % 2 == 0 and index < f.shape[0]-1 and row['Firing_Time']>0:
            firing_data.append(row['Firing_Time'])
            segmentLoc_data.append(row['Segment_location_of_firing'])
            waveDuration.append(row['Duration_of_wave(Frames)'])
            segmentsCovered.append(row['Number_of_Segments_covered'])
            waveDirection.append(row['Wave_direction'])
            Amplitude.append(row['Amplitude(median)'])
            
            
sorted_firing_data = sorted(firing_data)
sorted_segmentsCovered = sorted(segmentsCovered)
#print(sorted_firing_data)
#print(segmentLoc_data)
#print(waveDuration)
#print(sorted_segmentsCovered)
#print(waveDirection)
#print(Amplitude)
```

## Form dataframe with list as columns

```python
# Creating two categorical columns for each dataframe
Group1 = ['57F07-Gal4'] * 643 #create categorical list with same repeating values
Group2 = ['Control Group']* 669 #create categorical list with same repeating values

# Inhibition dataframe
InhibitionGroup_df = pd.DataFrame(
    {'FiringTime': firing_data,
     'SegmentLocationOfFiring': segmentLoc_data,
     'WaveDuration': waveDuration,
     'NumberOfSegmentsCovered': segmentsCovered,
     'DirectionOfWave': waveDirection,
     'Amplitude': Amplitude,
     'Group': Group1
    })

# Control dataframe
ControlGroup_df = pd.DataFrame(
    {'FiringTime': control_firing_data,
     'SegmentLocationOfFiring': control_segmentLoc_data,
     'WaveDuration': control_waveDuration,
     'NumberOfSegmentsCovered': control_segmentsCovered,
     'DirectionOfWave': control_waveDirection,
     'Amplitude': control_Amplitude,
     'Group': Group2
    })

# Combine both dataframes to a single one
C4daTargetsInhibition_df = InhibitionGroup_df.append(ControlGroup_df, ignore_index=True)
```
